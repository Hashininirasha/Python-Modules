```python
import numpy as np
x=np.arange(0,100,2)
print(x)
```

```python
y1=2*x+3
y2=np.power(x,2)
print(y1)
print(y2)
```

```python
plt.plot(x,y1,'r+',label="X vs Y1")
plt.plot(x,y2,'g.',label="X vs Y2")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.savefig('mygraph.png',dpi=300)
plt.show()
```
